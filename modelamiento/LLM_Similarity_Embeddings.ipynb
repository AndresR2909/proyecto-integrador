```python
# Embedding específico para comparar
reference_embedding = embeddings_text[42202]

dict_similarities = {}  # Diccionario para almacenar similitudes, norma Euclidiana y distancia de Manhattan

for index, embedding in zip(df_data.index.values, X):
    # Calcular similitud coseno con el embedding de referencia
    cosine_metric = cosine_similarity([embedding], [reference_embedding])[0][0]

    # Calcular norma Euclidiana y distancia de Manhattan
    euclidean_distance = np.linalg.norm(embedding - reference_embedding)
    manhattan_distance = np.linalg.norm(embedding - reference_embedding, ord=1)

    # Almacenar los valores en el diccionario si la similitud coseno es mayor o igual a 0.88
    if cosine_metric >= 0.88 or euclidean_distance <= 0.45 :
        dict_similarities[index] = {
            'cosine_similarity': cosine_metric,
            'euclidean_distance': euclidean_distance,
            'manhattan_distance': manhattan_distance
        }

# Mostrar el diccionario resultante
for key, value in dict_similarities.items():
    print(f"Índice: {key}")
    print(f"Similitud coseno: {value['cosine_similarity']}")
    print(f"Norma Euclidiana: {value['euclidean_distance']}")
    print(f"Distancia de Manhattan: {value['manhattan_distance']}")
    print(df_data["text"].iloc[key])
```
